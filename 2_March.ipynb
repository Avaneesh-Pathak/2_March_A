Q1: Matplotlib is a Python library used for creating static, animated, and interactive visualizations in Python. It provides a wide range of plotting functionality for creating various types of plots, charts, and graphs. Matplotlib is widely used in data analysis, scientific research, and data visualization tasks.

Five plots that can be created using the Pyplot module of Matplotlib are:

Line plot: Used to display data points connected by straight line segments.
Scatter plot: Shows the relationship between two variables as individual data points.
Bar plot: Represents categorical data with rectangular bars of different heights.
Histogram: Displays the distribution of a continuous variable by dividing it into bins and counting the frequency.
Pie chart: Represents data as slices of a circle, where the size of each slice corresponds to the proportion of the data it represents.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to display multiple plots in a single figure, arranged in rows and columns. The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the current plot.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title("Line 1")

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title("Line 2")

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title("Line 3")

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title("Line 4")

plt.tight_layout()
plt.show()
```

A bar plot, also known as a bar chart, is a visualization that represents categorical data using rectangular bars of different heights. The length or height of each bar corresponds to the quantity or value it represents. Bar plots are commonly used to compare and display data across different categories or groups.

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(8, 6))

# Bar plot
plt.subplot(1, 2, 1)
plt.bar(company, profit)
plt.title("Bar Plot")
plt.xlabel("Company")
plt.ylabel("Profit")

# Horizontal bar plot
plt.subplot(1, 2, 2)
plt.barh(company, profit)
plt.title("Horizontal Bar Plot")
plt.xlabel("Profit")
plt.ylabel("Company")

plt.tight_layout()
plt.show()
```

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the summary statistics of a dataset, such as the minimum, first quartile (25th percentile), median (second quartile or 50th percentile), third quartile (75th percentile), and maximum. The plot consists of a rectangular box (the interquartile range) and two whiskers (lines extending from the box) that represent the variability outside the box.

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.title("Box Plot")
plt.xlabel("Box")
plt.ylabel("Value")

plt.show()
```
